## 1.3 Organizing the images in a project

You can modify the thumbnail size, sort the images in the list by their metadata and add new key value pairs to the metadata.

From the context menu of the image list (right click on the name of the project or on an empty space in the list), change the thumbnail size to ``large``.

You can sort the images by name, id and uri (the path for local images). However in our example all three criteria result in the same order. Imagine that you have many images from an experiment with different conditions and controls. It could be useful to sort and group the images accordingly.

Imagine the first two images are from the condition ``ex1`` and the last image is a ``control``. 

* Select the first two images (you can use the usual ![Kenney-Ctrl](../../media/inputs-icons/keyboard_ctrl.svg)+![Kenney-Left-click](../../media/inputs-icons/mouse_left.svg) to select multiple entries or ![Kenney-Shift](../../media/inputs-icons/keyboard_shift.svg)+![Kenney-Left-click](../../media/inputs-icons/mouse_left.svg) to select a range)
* Add a new metadata field with the key ``condition`` and the value ``ex1`` to these images
* Select the third image
* Add a new metadata field with the key ``condition`` and the value ``control`` to these images
* Sort the images by ``condition``
