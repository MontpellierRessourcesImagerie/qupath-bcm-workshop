# 2. Navigation, Annotations and Measurements

In this section you will learn 
 - how to create annotations and modify their properties
 - about the different annotation types
 - how to create a hierachy of annotations
 - how to make measurements of annotations
 - how to use the magic wand tool
 - how to classify annotations
 - how to split annotations


## 2.1 Creating an annotation and setting its properties

Aims:
1. Create an annotation
1. Set its name, description and color
1. Modify the global display options for annotatins


### 2.1.1 Create a rectangular annotation

Open an image and switch to the annotations-tab. Select the rectangle tool from the toolbar and make a rectangular annotation somewhere on the image.

### 2.1.2 Set the properties of the annotation

Use ``Set properties`` from the context menu of the annotation to 
- set the *name*
- set the *description*
- change its *color*

<img src="../../media/qupath-navigation/annotation-properties.png" alt="annotation properties" >

**Note**

Note that the color will only become visible when the annotation is not selected anymore.
