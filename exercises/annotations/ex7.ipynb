## 2.7 Splitting Annotations

Aim:
- Split an annotation by a line

When annotations are the result of a pixel classifier, sometimes to separate objects that are close to each other can be merged together. You can split an annotation using a line. 

Make an annotation and a line across the annotation. Run the command ``Split annotations by line`` from the menu ``Objects>Annotations``.
