## 2.3 Annotation Hierarchy

Aim:
- Create a hierarchy of annotations

Make an area annotation. Then make two more area annotations which are comletely inside of the first annotation. Select the first annotation. From the ``more options``-button ![QP zoom-fit button](../../media/qp-icons/more-options.png) on the annotation list, run the command ``Insert in hieararchy`` ( ![Kenney-Ctrl](../../media/inputs-icons/keyboard_ctrl.svg) + ![Kenney-Shift](../../media/inputs-icons/keyboard_shift.svg) + ![Kenney-keyboard_i](../../media/inputs-icons/keyboard_i.svg) 
). Check the result on the hierarchy tab.

**Note**

You can use the command ``Resolve hierarchy`` (![Kenney-Ctrl](../../media/inputs-icons/keyboard_ctrl.svg) + ![Kenney-Shift](../../media/inputs-icons/keyboard_shift.svg) + ![Kenney-keyboard_r](../../media/inputs-icons/keyboard_r.svg)) from the menu ``Annotations`` to automatically build the hierarchy of all objects in the image.
