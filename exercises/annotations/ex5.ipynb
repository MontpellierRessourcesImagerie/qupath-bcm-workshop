## 2.5 The magic wand tool

Aim:
- Learn to use the magic wand tool
- Automatically fill holes in an annotation


### 2.5.1 Using the magic wand tool 

Open ``image012.lif`` from the ``01-multiplex-fluo`` dataset.

<center>
<img src="../../media/qupath-navigation/section.png" alt="the image for the exercise" >
<br>   
<small>

</small>
</center>

Use the wand tool ![QP zoom-fit button](../../media/qp-icons/wand-selection.png) to select the section on the slide.

**Tip**

1. Zoom out to quickly select big areas!
1. Zoom in to correct details!
1. Push from the inside with shift held down to enlarge the annotation!
1. Push from the outside with alt held down to shrink the annotation!

### 2.5.2 Fill holes

If the annotation you created is good at the contour but still has some holes on the inside, you can automatically fill them. With the annotation selected, run the command ``Fill holes`` from the menu ``Objects>Annotations``.

<center>
<img src="../../media/qupath-navigation/section-annotation.png" alt="annotation of the section" >
<br>   
<small>
Fig. 2.5.2 &ndash; An annotation of the section.
</small>
</center>

**Note**

The goal here is to learn how to use the magic wand tool. The annotation in this case could have been made more easily using a pixel classifier applying a simple threshold. You will learn how to do this in chapter 4.
