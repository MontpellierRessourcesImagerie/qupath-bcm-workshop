# Tissue classification/segmentation with QuPath

## Goals
These exercises aim to use different ways to locate some areas into some objects, or the objects themselves.

Examples:
- Find the objects on a slide
- Find the area of necrosis
- Find the area of positivity to a dye

Depending on our problem, we will want to produce either a new set of polygons corresponding to our areas of interest or a measure in µm² within the properties of a polygon.

## I. Fully manual: Glomerulus over kidney ratio

The most basic way to locate your objects is to manually create the polygons by yourself...
As you may expect, this approch is time-consuming, user dependend and may be nearly impossible depending what you need to find.
In this exercise we will try to process the ratio of glomerulus over the total area of kidney.
To do so, we will need to:

1. Find a polygon around each kidney
2. Find a polygon around the globerulus
3. Have the area of both
4. Export that into a spread sheet
5. Process the ratio
