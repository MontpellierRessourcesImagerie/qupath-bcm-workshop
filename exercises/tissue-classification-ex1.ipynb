# 4. Tissue segmentation and classification

## 4.1 Fully manual tissue segmentation

### 4.1.1 Glomerulus-to-kidney ratio

The most basic way to locate your objects is to manually create the polygons by yourself...
As you may expect, this approch is time-consuming, user dependend and may be nearly impossible depending what you need to find.
This technique can be used whether you need to locate your objects of interest, or sub-areas within them.

#### Goals:

- The goal of this exercise is to process the glomerulus-to-kidney ratio for every image you have.
- To do so, you will need to:
    - Find the polygon corresponding to the whole kidney and give it the `Kidney` classification.
    - Find the set of polygons representing glomerulus and give them the `Glomerulus` classification.
    - Repeat that on two or three images.
    - Export the measurements to a TSV so you can process the ratio per image in a spread sheet.
- After that, we will switch to some lungs images to have a more accurate idea of the limitations that you will encounter if you stick to fully manual segmentation.

### A. Locate the kidney

- On these images, kidneys are solid circular blocks with no to little holes (so we will ignore them).
- Before starting, make sure that you have a `Kidney` and a `Glomerulus` classes in your class list. If you miss them, you can generate them using the "+" button above the class list.
- To locate your kidneys, you can either use the ![ellipse button](../media/qp-icons/ellipse-selection.png) ellipse tool or the ![wand button](../media/qp-icons/wand-selection.png) wand tool. Don't forget that the scale of your wand depends on your zoom level. If you want to select the whole object, stay far!
- Like with any selection tool, you can inflate from inside or push from outside.
- Once you are done, before going to the next image, don't forget to:
    - assign the `Kidney` class to the annotation
    - Lock it with [Ctrl+K] or right-click > lock
    - Save the project
